## Missing Values

```python
# Print the info of df
df.info()
```

```python
df.describe()
```

`.describe()` can only be used on numeric columns. So how can you diagnose data issues when you have categorical data? One way is by using the `.value_counts()` method, which returns the frequency counts for each unique value in a column!

You want to set the `dropna` column to `False` so if there are missing values in a column, it will give you the frequency counts.

```python
# Print the value counts for 'Borough'
print(df['Borough'].value_counts(dropna=False))

# Print the value_counts for 'State'
print(df['State'].value_counts(dropna=False))

# Print the value counts for 'Site Fill'
print(df['Site Fill'].value_counts(dropna=False))
```

## Visualizing and identifying outliers

* Histograms are great ways of visualizing single variables.

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Plot the histogram
df['Existing Zoning Sqft'].plot(kind='hist', rot=70, logx=True, logy=True)

# Display the histogram
plt.show()
```

The plot shows us that there are barely any counts near the `max` value, signifying an outlier.

## Visualizing multiple variables with boxplots

* To visualize multiple variables, boxplots are useful, especially when one of the variables is categorical.

```python
# Cleaning data
df['initial_cost'] = df['Initial Cost'].apply(lambda x: float(x.split('$')[1]))

# Create the boxplot of initial_cost accross different values of Borough
df.boxplot(column='initial_cost', by='Borough', rot=90)

# Display the plot
plt.show()
```

You can see the 2 extreme outliers are in the borough of Manhattan. An initial guess could be that since land in Manhattan is extremely expensive, these outliers may be valid data points. Again, further investigation is needed to determine whether or not you can drop or keep those points in your data.

```python
# Cleaning data
df['total_est_fee'] = df['Total Est. Fee'].apply(lambda x: float(x.split('$')[1]))

# Create
df.plot(kind='scatter', x='initial_cost', y='total_est_fee', rot=70)
plt.show()
```
