# Seaborn Introduction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/schoolimprovement2010grants.csv')
```

## Comparing a histogram and distplot - Univariate Analysis
The pandas library supports simple plotting of data, which is very convenient when data is already likely to be in a pandas DataFrame.

Seaborn generally does more statistical analysis on data and can provide more sophisticated insight into the data.

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()
```

```python
# Display a Seaborn distplot
sns.distplot(df['Award_Amount'])
plt.show()

# Clear the distplot
plt.clf()
```

By disabling the KDE output, you can create a histogram.

```python
# Create a distplot
sns.distplot(df['Award_Amount'],
             kde=False,
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a distplot of the Award Amount
sns.distplot(df['Award_Amount'],
             hist=False,
             rug=True,
             kde_kws={'shade':True}) # keyword agruments for kde_plot()

# Plot the results
plt.show()
```

## Regression Plots

```python
df = pd.read_csv('data/insurance_premiums.csv')
```

```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(x='insurance_losses',
            y='premiums',
            data=df)


# Display the plot
plt.show()
```

* The solid line is evidently the linear regression model fit.
* The translucent band lines, however, describe a bootstrap confidence interval generated for the estimate. From the [docs](https://seaborn.pydata.org/generated/seaborn.regplot.html) we can see this in the parameter info for `ci`:
> Size of the confidence interval for the regression estimate. This will be drawn using translucent bands around the regression line. The confidence interval is estimated using a bootstrap; for large datasets, it may be advisable to avoid that computation by setting this parameter to None.

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(x='insurance_losses',
           y='premiums',
           data=df)


# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```
