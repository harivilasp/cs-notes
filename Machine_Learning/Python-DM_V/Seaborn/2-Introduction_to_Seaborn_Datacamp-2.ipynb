# Introduction to Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Customizing scatter plots

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            size="size")
plt.show()
```

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            size="size",
            hue="size")
plt.show()
```

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            hue="smoker",
            style="smoker")
plt.show()
```

```python
# Set alpha to be between 0 and 1
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            alpha=0.4)
plt.show()
```

## Interpreting line plots

```python
mpg = pd.read_csv('data/mpg.csv')
```

```python
# Create line plot
sns.relplot(x = 'model_year', y = 'mpg', data = mpg, kind = 'line')

# Show plot
plt.show()
```

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci = 'sd')

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", markers = True,dashes = False,
            hue="origin")

# Show plot
plt.show()
```

## Count plots and bar plots

```python
sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar")
plt.show()
```

## Box Plots

```python
student_data = pd.read_csv('data/student_data.csv')
```

```python
student_data.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
student_data.columns
```

```python
student_data.head()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x = 'study_time', y = 'G3', data = student_data, order = study_time_order, kind = 'box')

# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x = 'internet',  y = 'G3', data = student_data, hue = 'location', kind = 'box', sym = '')

# Show plot
plt.show()
```

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis = 2.0)

# Show plot
plt.show()
```

```python
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])

# Show plot
plt.show()
```
