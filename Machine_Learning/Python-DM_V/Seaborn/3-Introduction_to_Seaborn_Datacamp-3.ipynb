```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Customizing point plots

### What are point plots?

* Points show mean of quantitative variable
* Vertical lines show 95% condence intervals

### Point plots vs. line plots

* Both show:
    * Mean of quantitative variable
    * 95% condence intervals for the mean
* Differences:
    * Line plot has **quantitative** variable (usually time) on x-axis
    * Point plot has **categorical** variable on x-axis
    
### Point plots vs. bar plots

* Both show:
    * Mean of quantitative variable
    * 95% condence intervals for the mean
* Differences:
    * Usually easier to compare differences on a point plot

```python
sns.catplot(x="smoker",
            y="total_bill",
            hue='sex',
            data=tips,
            kind="point")
plt.show()
```

```python
sns.catplot(x="smoker",
            y="total_bill",
            data=tips,
            kind="point",
            estimator=np.median,
            join=False,
            capsize=0.2)
plt.show()
```

```python
student_data = pd.read_csv('data/student_data.csv')
```

```python
# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator=np.median)

# Show plot
plt.show()
```

## Changing style and palette

```python
mpg = pd.read_csv('data/mpg.csv')
```

```python
# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu_r")
sns.set_context("talk")

sns.relplot(x="horsepower",
            y="weight",
            data=mpg,
            hue="mpg",
            kind="scatter")

# Show plot
plt.show()
```

## FacetGrids vs. AxesSubplots

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation = 90)

# Show plot
plt.show()
```
